```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def timeToDecimal(a):
    x = 0
    y = 0
    if len(a) == 4:
        x = float(a[0:1])
        y = float(a[2:])
        
    if len(a) == 5:
        x = float(a[0:2])
        y = float(a[3:])
    x *= 60
    z = (x+y)/1440
    return z
```

```python
data = np.genfromtxt('ASTR19_F24_group_project_data.txt', dtype=[('day', 'i8'), ('time', 'U6'),('height', 'f8')])

day = []
time = []
height = []

for i in range (82):
    day.append(data[i][0])
    time.append(data[i][1])
    height.append(data[i][2])

real_Day = []
for i in range (len(time)):
    x = day[i] + timeToDecimal(time[i])
    x_1 = round(x, 3)
    real_Day.append(x_1)
```

```python
plt.plot(real_Day, height, '-o')
plt.xlabel("Day")
plt.ylabel("Tide height")
plt.title("Tide heights over time")
plt.savefig("Tide heights over time.pdf")
plt.show()
```

```python
#the data to be inserted for day 14
insert_day = 14
insert_time = "23:47"
insert_height = 7.33

#how index is calculated based on day values
insert_index = len(day) #defaults to end of array
for i in range(len(day)):
    if day[i] > insert_day:
        insert_index = i
        break

#inserts the new data
day.insert(insert_index, insert_day)
time.insert(insert_index, insert_time)
height.insert(insert_index, insert_height)

#recalulates real day so that time and day are matched together
real_Day = []
for i in range(len(time)):
    x = day[i] +timeToDecimal(time[i])
    x_1 = round(x, 3)
    real_Day.append(x_1)

plt.plot(real_Day, height, '-o')
plt.xlabel("Day")
plt.ylabel("Tide height")
plt.title("Tide heights over time")
plt.savefig("Updated Graph of Tide heights over time.pdf")
plt.show()
```
